```python
import cv2
import matplotlib.pyplot as plt
from pathlib import Path
from king_tensor import Screen
```

```python
%matplotlib inline
```

```python
root_dir = Path().resolve().parents[0]
root_dir
```

```python
screen_path = root_dir / 'data' / 'IMG_1470.PNG'
img = cv2.imread(screen_path, cv2.IMREAD_GRAYSCALE)
plt.imshow(img)
plt.show()
```

```python
screen = Screen(img, root_dir/'king_tensor'/'config'/'ui.yaml')
elixir = screen.images['elixir_bar']
plt.imshow(elixir)
plt.show()
```

```python
ret, thresh = cv2.threshold(elixir, 75, 255, cv2.THRESH_BINARY)
plt.imshow(thresh)
plt.show()
```

We can try cropping further to isolate a very small region of the meter

```python
crop = thresh[45:65, 83:1118]
plt.imshow(crop)
plt.show()
```
