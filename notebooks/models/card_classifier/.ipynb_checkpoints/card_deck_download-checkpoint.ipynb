```python
import requests
from pathlib import Path
import re
```

```python
def download_png(image_url, save_path):
    
    try:
        # Send a GET request to the URL
        response = requests.get(image_url)
        # Raise an exception if the request was unsuccessful
        response.raise_for_status() 

        # Open a file in write-binary mode ('wb') and write the content
        with open(save_path, 'wb') as file:
            file.write(response.content)
        print(f"Image downloaded successfully to {save_path}")

    except requests.exceptions.RequestException as e:
        print(f"Failed to download image: {e}")
```

```python
deck_url_path = Path() / '../data/deckshop_html.txt'
```

```python
with open(deck_url_path, 'r') as file:
    lines = file.readlines()
```

```python
evo_pattern = re.compile(
    r'<a href="\/card\/detail\/(.+?)\?evolution=(0|1)">'
    r'<img src="\/img\/(.+?)\/(.+?)\.png" alt="(.+?)" class="card"><\/a>'
)
pattern = re.compile(
    r'<a href="\/card\/detail\/(.+?)">'
    r'<img src="\/img\/card_ed\/(.+?)\.png" alt="(.+?)" class="card"><\/a>'
)

_ = '<a href="/card/detail/little-prince"><img src="/img/card_ed/LittlePrince.png" alt="Little Prince" class="card"></a'

evo_lines = []
norm_lines = []
for line in lines:
    line = line.strip()
    evo_match = evo_pattern.fullmatch(line)
    norm_match = pattern.fullmatch(line)
    if evo_match and norm_match:
        evo_lines.append(line)
    elif norm_match:
        norm_lines.append(line)

norm_lines = list(set(norm_lines))
```

```python
base_url = 'https://www.deckshop.pro'
out_dir = Path() / '../data/deckshop_cards/'
```

```python
for line in norm_lines:
    img_url = line.split('"')[3]
    url = base_url + img_url
    card_name = url.split('/')[-1].strip('.png')
    download_png(url, out_dir / (card_name + '.png'))
    
for line in evo_lines:
    img_url = line.split('"')[3]
    url = base_url + img_url
    card_name = url.split('/')[-1].strip('.png')
    download_png(url, out_dir / ('evo_' + card_name + '.png'))
```
